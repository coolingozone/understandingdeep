## Simple Example for backpropagation

```python
#import the matplotlib for ploting graph
import matplotlib.pyplot as plt

#initialize all the parameters
x=0.1
y=0.5
w1=0.5
w2=0.9
w3=0.8
zt=0.7
ae=[]
#initialize learning rate
learnrate=0.01
n=100
#train for n number of epoch
for i in range(0,n):
  q=w1*x+w2*y
  z=q*w3
  e=zt-z
  ae.append(e*e)
  dw3=-q*e
  dw2=e*w3*y
  dw1=e*w3*x
  w3=w3-learnrate*dw3
  w2=w2-learnrate*dw2
  w1=w2-learnrate*dw1

plt.plot(range(0,n),ae)
plt.ylabel('Squre Error')
plt.xlabel('Epoch')
plt.show()


```
